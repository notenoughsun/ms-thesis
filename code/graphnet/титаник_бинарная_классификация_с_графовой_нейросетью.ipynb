```python
import pandas as pd
from stellargraph import StellarGraph
```

```python
train = pd.read_csv("train.csv")
train
```

```python
train = train.drop(["Name","Sex","Ticket","Cabin","Embarked"], axis=1)
```

```python
train = train.drop(["Age"], axis=1)
```

```python
train.to_csv("titanic_clean.csv", index=False)
```

# Данные титаника представляем в графовом виде

```python
cora_cites = train[["PassengerId","Pclass"]]
cora_cites
```

```python
cora_cites.to_csv("cora_cites.csv", index=False)
```

```python
cora_raw_content = train[["PassengerId","SibSp","Parch","Fare","Survived"]]
cora_raw_content
```

```python
cora_content_str_subject = cora_raw_content.set_index("PassengerId")
cora_content_str_subject
```

```python
cora_content_str_subject.to_csv("cora_content_str_subject.csv", index=True)
```

# Запуск графовой нейросети

```python
from sklearn import model_selection

cora_train, cora_test = model_selection.train_test_split(
    cora_content_str_subject, train_size=0.25, random_state=123
)
cora_train
```
